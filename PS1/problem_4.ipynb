```python
import cudaq 
import os
import numpy as np

cudaq.set_target("qpp-cpu")
cudaq.set_random_seed(0)
```

```python
os.makedirs("PS1/results", exist_ok=True)

@cudaq.kernel
def rotate_x(theta: float):
    q = cudaq.qubit()
    rx(theta, q) # Rotate the qubit around the X-axis by angle theta


blochSphereList = []
for theta in [np.pi/4, np.pi/2, 3*np.pi/4, np.pi]:
    sph = cudaq.add_to_bloch_sphere(cudaq.get_state(rotate_x, theta))
    blochSphereList.append(sph)

cudaq.show(blochSphereList[:4], nrows=2, ncols=2)
```
